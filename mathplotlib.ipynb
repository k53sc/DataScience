# mathplotlib Introduction

```python
# We inport the package in this way
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# Lets create somoe data that will be plotted
x = [ 1, 2, 3 ]
y = [ 4, 5, 6 ]
```

```python
# plot the graph
plt.plot( x , y )

# specify the range of the axis
plt.axis( [ 0, 10, 0, 10 ] )
```

```python
# jif we have x axis data as string then we have to map it to integers then plot
names = [ 'a', 'b', 'c']
x = range(3)
plt.xticks(x, names)
plt.plot(x,y,"g")
plt.show()
```

```python
# multiple graphs in one
# g and b stand for color
x = [ 1, 2, 3 ]
y = [ 4, 5, 6 ]
y1 = [ 5, 6, 7 ]
plt.plot(x,y,"g")
plt.plot(x,y1,"b" )
plt.show()
```

```python
# controlling line width
x = [ 1, 2, 3 ]
y = [ 4, 5, 6 ]
y1 = [ 5, 6, 7 ]
plt.plot(x,y,"g", linewidth=1.0)
plt.plot(x,y1,"b", linewidth=4.0)
plt.show()
```
